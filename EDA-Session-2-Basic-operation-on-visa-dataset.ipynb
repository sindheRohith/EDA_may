**Task-1**

- i want to create two lists
- one is categorical column list
- second one numerical column list
- using dtypes

- whenever you want to create any list from series
- first convert the series into a dictionary
- when you convert it becomes akey value pair
- iterate through dictionary
- if that value == onject type, then keep that corresponding key in a one list
- esle keep that variable in other list

```python
data_types=visa_df.dtypes
dict(data_types)
```

```python
cat_list=[]
num_list=[]
for i,j in dict(data_types).items():
    if j=='object':
        cat_list.append(i)
    else:
        num_list.append(i)
    print(i)
```

```python
cat_list=[i for i,j in dict(data_types).items() if j=='object']
num_list=[i for i,j in dict(data_types).items() if j!='object']
cat_list,num_list
```

```python
len(cat_list),len(num_list)
```

- we can not create a dataframe using these two lists
- because these two list are not in equal lenght
- if you want to create a dataframe , then we need to make two list has equal lenght
   - then you need to question yourself
   - making equal lenght required or not
   - if it is require how?

**create columns dataframe**

```python
num_df=pd.DataFrame(num_list,columns=['Numerical_column'])
num_df.to_csv('Numerical_column_df.csv',index=False)
num_df
```

```python
num_df=pd.DataFrame(num_list,columns=['categorical_column'])
num_df.to_csv('categorical_column_df.csv',index=False)
num_df
```
